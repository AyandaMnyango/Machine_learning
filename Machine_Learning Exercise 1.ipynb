```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures


#Setting the range of random numbers generated.
np.random.seed(5)

#Setting the x values of the graph.
X = np.linspace(-5, 5, 80).reshape(-1, 1)
print(X.shape)

#Applying a cosine function on every X value.
y_variable = np.cos(1.5*X).ravel()#Structures the values into a 1D array


#mean, std, len(X)
y = y_variable + np.random.normal(0, 0.25, len(X))

#Fit linear regression model
model = LinearRegression()
model.fit(X,y)
y_linear = model.predict(X)

#Fit a degree -4 polynomial model and explain why it fits better.
#y = a0 + a1x + a2x^2 + a3x^3 + a4x^4

poly4 = PolynomialFeatures(degree=4, include_bias=False)
X_poly4 = poly4.fit_transform(X)

#Create object of degree 4.

poly4_model = LinearRegression()
poly4_model.fit(X_poly4, y)
y_poly4 = poly4_model.predict(X_poly4)


plt.figure()
plt.scatter(X,y, label="Data")
plt.plot(X,y_linear, label="underfitting")
plt.plot(X, y_poly4, label="Good fit")
plt.title("Overfitting vs Underfitting")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()

#Create object of degree 12.
#Fit a degree -12 polynomial model and explain why it fits better.
#y = a0 + a1x + a2x^2 + a3x^3 + a4x^4 + a5x^5 + a6x^6 + a7x^7 + a8x^8 + a9x^9
#+ a10x^10 + a11x^11 + a12x^12plt.plot(X, y_poly4, label="Degree 4")

poly12 = PolynomialFeatures(degree=12, include_bias=False)
X_poly12 = poly12.fit_transform(X)


poly12_model = LinearRegression()
poly12_model.fit(X_poly12, y)
y_poly12 = poly12_model.predict(X_poly12)

plt.figure()
plt.scatter(X,y, label="Data")
plt.plot(X,y_linear, label="Overfit")
plt.plot(X, y_poly12, label="Degree 12")
plt.title("Overfitting vs Underfitting")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures


#Setting the random generated values to 7.
np.random.seed(7)

X = np.linspace(-3, 3, 80).reshape(-1,1)
y_true = np.sin(2 * X).ravel()

y = y_true + np.random.normal(0, 0.3, len(X))

#Inserting the Linear Regression.
model = LinearRegression()
model.fit(X,y)
y_linear = model.predict(X)

#plot linear Regression.
plt.figure()
plt.scatter(X, y_true, label="Data")
plt.plot(X, y_linear, label="Linear Regression")
plt.title("Underfitting vs Overfitting")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()

#Insering Polynomial Features
poly3 = PolynomialFeatures(degree=3, include_bias=False)
X_poly3 = poly3.fit_transform(X)
model_poly3 = LinearRegression()
model_poly3.fit(X_poly3, y)

#Ploting the graph
plt.figure()
plt.scatter(X, y, label="Data")
plt.plot(X, y_linear, label="Underfit")
plt.plot(X, model_poly3.predict(X_poly3), label="Degree 3")
plt.title("Overfitting vs Underfitting")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()

#Creating Degree 10.
poly10 = PolynomialFeatures(degree=10, include_bias=False)
X_poly10 = poly10.fit_transform(X)
model_poly10 = LinearRegression()
model_poly10.fit(X_poly10,y)

#Ploting the graph for Degree 10.
plt.figure()
plt.scatter(X,y, label="Data")
plt.plot(X, y_linear, label="Underfit")
plt.plot(X, model_poly10.predict(X_poly10), label="Degree 10")
plt.title("Overfitting vs Underfitting")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


#Creating a given dataset.
Data = {
    "Age": [22, 25, np.nan, 35, 40, 40, 28, np.nan, 30, 30],
    "Income":[3000, 3500, 4000, np.nan, 8000, 8000, 4200, 5000, np.nan, np.nan],
    "Score": [0.2, 0.4, 0.35, 0.8, 0.9, 0.9, 0.5, 0.55, 0.7, 0.7],
    "Bought": [0, 0, 0, 1, 1, 1, 0, 0, 1, 1]
}
df = pd.DataFrame(Data)
print(df)

#Finding Missing Values.
print(df.isnull().sum())

#Finding Duplicated values in the table.
print(f"There are {df.duplicated().sum()} duplicated values.")

#Removing Duplicated Values.
df = df.drop_duplicates()
print(df)

#Filling in missing values with a mean.

df["Age"] = df["Age"].fillna(df["Age"].mean())
df["Income"] = df["Income"].fillna(df["Income"].mean()).round(2)

print(df)

from sklearn.preprocessing import MinMaxScaler, StandardScaler

#Using Standard Scaler.
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df)
print(df_scaled)
```
