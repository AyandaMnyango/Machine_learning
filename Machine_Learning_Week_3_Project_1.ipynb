```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

#Show how a tree fits data in pieces (step-like predictions) and can overfit if too deep. Print MSE and R^2.

n = 100
np.random.seed(8)
X = np.random.uniform(0, 10, n).reshape(-1, 1) # (n_sample, n_features)
y = np.sin(X).ravel() + np.random.normal(0, 0.23, n) #(n_sample)

X_train, X_test, y_train, y_test = train_test_split(X,y, train_size=0.25, random_state=35)

shallow_tree = DecisionTreeRegressor(max_depth=3, random_state=42)
shallow_tree.fit(X_train, y_train)
y_predict_shallow = shallow_tree.predict(X_test)

deep_tree = DecisionTreeRegressor(max_depth=None, random_state=42)
deep_tree.fit(X_train, y_train)
y_predict_deep = shallow_tree.predict(X_test)

mse_shallow_tree = mean_squared_error(y_test, y_predict_shallow)
r2_shallow = r2_score(y_test, y_predict_shallow)
print(f"The MSE for shallow tree {round(mse_shallow_tree, 3)}")
print(f"The R2 for shallow tree {round(r2_shallow, 3)}")

print("\n############## Deep tree ##############")
mse_deep_tree = mean_squared_error(y_test, y_predict_deep)
r2_deep_tree = r2_score(y_test, y_predict_deep)

print(f"The MSE for deep tree {round(mse_deep_tree), 3}")
print(f"The R2 for deep tree {round(r2_deep_tree, 3)}")

x_even_spaced = np.linspace(0, 20, 600).reshape(-1,1)
plt.figure()
plt.scatter(X, y, label="Data")
plt.plot(x_even_spaced, shallow_tree.predict(x_even_spaced), label="Shallow")
plt.plot(x_even_spaced, deep_tree.predict(x_even_spaced), label="deep tree")
plt.title("Deep vs shallow tree")
plt.show()
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

np.random.seed(8)
size = np.random.uniform(2, 50, 100)
noise = np.random.normal(0, 200, 100)
price = size + noise
#
df = pd.DataFrame({"Size": size, "price": price})
X = df[["Size"]]
y = df["price"]

#X_train
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=30)
model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Intercept {model.intercept_}")
print(f"The slope is {model.coef_[0]}")
print(f"The mean square error {round(mse, 2)}")
print(f"The r2 {round(r2, 2)}")
```
