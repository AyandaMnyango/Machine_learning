```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error, r2_score


np.random.seed(4)
x = np.random.uniform(0, 20, 100).reshape(-1, 1)
y = np.cos(x).ravel + np.random.normal(0, 0.25, 100)

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=30)

#linear model
linear_model = LinearRegression()
linear_model.fit(x_train, y_train)
y_lin_predict = linear_model.predict(x_test)
mse_linear = mean_squared_error(y_test, y_lin_predict)


#Polynomial
poly7 = PolynomialFeatures(degree=7, include_bias=False)
x_train_poly7 = poly7.fit_transform(x_train)
```
