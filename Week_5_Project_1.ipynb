```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay


X, y = load_breast_cancer(return_X_y=True)

X_train, X_test, y_train, y_test = train_test_split(X,y, train_size=0.25, random_state = 30)

model = LogisticRegression()
model.fit(X_train, y_train)
y_predict = model.predict(X_test)
accuracy = accuracy_score(y_test, y_predict)
confusion_matrix = confusion_matrix(y_test, y_predict)

print(f"The Accuracy is : {accuracy :.2f}")
print(f"The confusion matrix is: {confusion_matrix}")

ConfusionMatrixDisplay.from_predictions(y_test, y_predict)
plt.title("The confusion matrix")
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier

X,y = load_iris(return_X_y=True)

X_train, X_test, y_train, y_test = train_test_split(X,y, train_size=0.25, random_state = 30)

model = LogisticRegression()
model.fit(X_train, y_train)
y_predict = model.predict(X_test)

#The Kneighbors Classifier for K = 1
model = KNeighborsClassifier(n_neighbors=1)
model.fit(X_train, y_train)
y_predict1 = model.predict(X_test)
accuracy1 = accuracy_score(y_test, y_predict1)

#The Kneighbors Classifier for K = 5
model = KNeighborsClassifier(n_neighbors=5)
model.fit(X_train, y_train)
y_predict5 = model.predict(X_test)
accuracy5 = accuracy_score(y_test, y_predict5)

#The KNeighbors Classifier for K = 10
model = KNeighborsClassifier(n_neighbors=10)
model.fit(X_train, y_train)
y_predict10 = model.predict(X_test)
accuracy10 = accuracy_score(y_test, y_predict10)

print(f"The accuracy for K = 1 is {accuracy1 :.2f}")
print(f"The accuracy for K = 5 is {accuracy5 :.2f}")
print(f"The accuracy for K = 10 is {accuracy10 :.2f}")
```
